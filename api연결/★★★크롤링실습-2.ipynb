```python
#########################################
#### [작업2]
####  교재 312페이지 변형
########################################

import urllib.request
from bs4 import BeautifulSoup
import pandas as pd

url='https://beomi.github.io/beomi.github.io_old/'
sorce_code=urllib.request.urlopen(url)
plain_text=sorce_code.read()

convert_data=BeautifulSoup(plain_text,'html.parser')

title=[x.string for x in convert_data.select('a')]

date=[]
for x in convert_data.select('span'):
    date.append(x.string)

urlList=[]
for x in convert_data.select('a'):
    urlList.append('https://beomi.github.io'+x.get('href'))


df=pd.DataFrame({'제목':title,
                '작성일':date,
                'url':urlList})
df.to_csv('beomi.csv',encoding='utf-8')
df.head()
```
