```python
import requests
import pandas as pd

def makeUrl(findAPI):
    try:
        apiEndPoint=dict({'':'blog',
                    '':'news',
                    '':'shop',
                    '':'book',
                    '':'movie', # 23 
                    '': 'kin',
                    '':'local',
                    '':'cafearticle',
                    '':'encyc',
                    '':'doc'})
        makeAPI=apiEndPoint[findAPI]
        endPointUrl=f'https://openapi.naver.com/v1/search/{makeAPI}.json'
        print(endPointUrl)
        return endPointUrl


    except:
       print(' {findAPI} , //////   ')


def naver_api(url, query,display=10):
    client_id = 'UGWFRw8nJK3z1B_irpX9'
    client_secret = 'tGq_GW6jBz'

    #     .
    headers = {
        'X-Naver-Client-Id': client_id,
        'X-Naver-Client-Secret': client_secret
    }

    params = {
        "query": query,
        "display": display  #    ( 100)
    }

    # API GET  .
    response = requests.get(url, headers=headers, params=params)

    if response.status_code == 200:
        result = response.json()
        df = pd.DataFrame(result['items'])
        return df
    else:
        print("Error:", response.status_code)
        return None


api,query='',''
url=makeUrl(api)
df=naver_api(url,query,100)
df
```

```python
df.to_csv(f' {api} api {query} .csv', index=False, encoding='euc-kr')
```
