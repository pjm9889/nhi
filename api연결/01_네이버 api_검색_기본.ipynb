### [작업1] 네이버 쇼핑몰 A
- url = 'https://openapi.naver.com/v1/search/shop.json' # 쇼핑몰
- response 응답번호를 정리합니다.

```python
import requests

# 네이버 개발자 센터에서 발급받은 클라이언트 ID와 시크릿을 입력합니다.
client_id = 'UGWFRw8nJK3z1B_irpX9'
client_secret = 'tGq_GW6jBz'
query = '노트북'

url = 'https://openapi.naver.com/v1/search/shop.json' # 쇼핑몰


# 요청 헤더에 인증 정보를 추가합니다.
headers = {
    'X-Naver-Client-Id': client_id,
    'X-Naver-Client-Secret': client_secret
}

params = {
    "query": query,
    "display": 10  # 가져올 결과 수 (최대 100)
}

# API에 GET 요청을 보냅니다.
# 구글검색 '파이썬 response 응답번호'
response = requests.get(url, headers=headers,params=params)
response, response.status_code
```

#### [작업3] 코드정리
- 함수화 하여서 코드를 정리합니다.

```python
import requests
import pandas as pd

def makeUrl(findAPI):
    try:
        apiEndPoint=dict({'블로그':'blog',
                    '뉴스':'news',
                    '쇼핑':'shop',
                    '책':'book',
                    '영화':'movie', # 23년 서비스종료
                    '지식': 'kin',
                    '지역':'local',
                    '카페':'cafearticle',
                    '지식백과':'encyc',
                    '학술정보':'doc'})
        makeAPI=apiEndPoint[findAPI]
        endPointUrl=f'https://openapi.naver.com/v1/search/{makeAPI}.json'
        print(endPointUrl)
        return endPointUrl


    except:
       print('요청한 {findAPI} 없음, 뉴스/쇼핑/책/영화/지식/지역/카페  만 가능')


def naver_api(url, query,display=10):
    client_id = 'UGWFRw8nJK3z1B_irpX9'
    client_secret = 'tGq_GW6jBz'

    # 요청 헤더에 인증 정보를 추가합니다.
    headers = {
        'X-Naver-Client-Id': client_id,
        'X-Naver-Client-Secret': client_secret
    }

    params = {
        "query": query,
        "display": display  # 가져올 결과 수 (최대 100)
    }

    # API에 GET 요청을 보냅니다.
    response = requests.get(url, headers=headers, params=params)

    if response.status_code == 200:
        result = response.json()
        df = pd.DataFrame(result['items'])
        return df
    else:
        print("Error:", response.status_code)
        return None


api,query='책','파이썬'
url=makeUrl(api)
df=naver_api(url,query,100)
df
```

```python
df.to_csv(f'네이버 {api} api {query} 검색결과.csv', index=False, encoding='euc-kr')
```
