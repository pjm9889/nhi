###문제2 - 파이썬 IDLE, 주피터 노트북, 구글 코랩 각각의 실습 환경 특징을 작성하시오.

#### 1) 파이썬 IDLE
파이썬 홈페이지(https://www.python.org/)에서 설차 파일을 다운로드 받아 설치한다. 파이썬 IDLE는 셀 모드와 편집 모드의 두 가지 종류로 구성된다.

#### 2) 주피터 노트북
아나콘다 홈페이지(https://www.anaconda.com/products/distribution)에서 설차 파일을 다운로드 받아 설치한다. 주피터 노트북 홈페이지에 화면 오른쪽 상단에 위치한 [New]를 클릭한 뒤, [Python 3 (ipykernel)]을 선택하여 소스코드를 작성한다.

#### 3) 구글 코랩
별도의 설치없이 코랩 사이트(https://colab.research.google.com/)에 방문하면 코랩에서 최근 작업한 파일 목록이 나타난다. 새로운 작업을 위해 화면 오른쪽 하단에 [새 노트]를 클릭한다. 또는 [취소] 버튼을 클릭한 후 나타나는 화면에서 [파일] → [새 노트]를 클릭하여 소스코드를 작성한다.

###문제3 - 데이터 분석을 위한 파이썬 IDLE, 주피터 노트북, 구글 코랩 각각의 실습 환경 특징을 작성하시오.

#### 1) 파이썬 IDLE
파이썬 IDLE에서 데이터 분석에 필요한 라이브러리를 사용하기 위해서는 파이썬과 별도로 각 라이브러리들마다 다운로드 및 설치 과정을 거쳐야 한다. 이 과정은 생각보다 복잡하고 버전 관리가 쉽지 않다는 단점이 있다.

#### 2) 주피터 노트북
주피터 노트북(Jupyter Notebook)은 웹 브라우저에서 파이썬 코드를 작성하고 실행할 수 있는 오픈소스 기반의 웹 플랫폼 개발 환경이다. 주피터 노트북은 데이터 분석에 유용하게 사용할 수 있는데, 특히 머신러닝과 딥러닝에 많이 활용되고 있다. 또한 주피터 노트북은 다양한 프로그래밍 언어 지원, 데이터 분석에 유용한 소스코드 공유 및 시각화 등의 특징을 가지고 있다. 이와 같은 장점에도 불구하고 다른 디바이스에서 같은 프로젝트를 이어서 진행하기 비교적 까다롭다는 단점이 있다.

#### 3) 구글 코랩
코랩(Colab)에서 코드 실행은 구글 클라우드 가상 서버에서 이루어지고, 파이썬이나 파이썬 패키지를 설치하지 않아도 사용할 수 있으며, 스마트폰으로 접속하여 코드를 실행할 수 있다. 코랩을 통해 진행 상황과 결과를 다른 사람들에게 편하게 공유할 수 있으며, 초보자들도 설치 과정의 어려움 없이 코드를 실행할 수 있다. 코랩에서 만든 파일은 구글 드라이브에 저장하여 불러올 수 있으며, 구글 드라이브 문서와 같이 링크로 접근하여 협업이 가능하다. 단, 구글 계정이 필요하고 인터넷이 연결되는 환경이어야 하며, 코드 실행 시 딜레이가 발생할 수 있다.
