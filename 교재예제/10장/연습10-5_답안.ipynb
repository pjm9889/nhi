```python
df=pd.read_csv('/content/KR_youtube_trending.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.publishedAt=pd.to_datetime(df.publishedAt,format='%Y-%m-%dT%H:%M:%SZ')
df.info()
```

```python
df.trending_date=pd.to_datetime(df.trending_date,format='%Y-%m-%dT%H:%M:%SZ')
df.info()
```

```python
def getDatetime(x):
  if 4<x.hour<=10:
    return 'Morning'
  elif 10<x.hour<=17:
    return 'Afternoon'
  elif 17<x.hour<=22:
    return 'Evening'
  else:
    return 'Night'

df['datetime']=df.publishedAt.apply(lambda x:getDatetime(x))
df.head()
```

```python
df_groupby=df.groupby('datetime').agg({'datetime':'count','likes':'mean','dislikes':'mean'})
df_groupby.head()
```

```python
sns.barplot(x=df_groupby.index,y='datetime',data=df_groupby)
```

```python
sns.barplot(x=df_groupby.index, y='likes',data=df_groupby)
```

```python
sns.barplot(x=df_groupby.index, y='dislikes',data=df_groupby)
```

```python
df['time']=df['trending_date']-df['publishedAt']
df.info()
```

```python
plt.figure(figsize=(15,10))
df['time'].value_counts().plot(label='인기동영상이 되기까지 걸리는 시간')
```

```python
plt.figure(figsize=(15,10))
sns.countplot(y='time',data=df)
```
