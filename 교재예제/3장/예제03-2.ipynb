## 3.9 파일 입출력

```python
f=open('name.txt')
data=f.read()
print(data)
f.close()
```

```python
f=open('name.txt')
data=f.readline()
print(data)
f.close()
```

```python
f=open('name.txt')
while True:
    data=f.readline()
    if not data:break
    print(data,end='')
f.close()
```

```python
f=open('name.txt')
data=f.readlines()
print(data,end='')
f.close()
```

```python
f=open('name.txt')
data=f.readlines()
for line in data:
    print(line,end='')
f.close()
```

```python
with open('name.txt') as f:
    data=f.readlines()
    print(data,end='')
```

```python
f=open('club.txt','w')

while True:
    name=input('이름: ')
    if not name:break
    f.write(name)
f.close()
```

```python
f=open('club.txt','w')

while True:
    name=input('이름: ')
    if not name:break
    f.write(name+'\n')
f.close()
```

```python
f=open('club.txt','a')

while True:
    name=input('이름: ')
    if not name:break
    f.write(name+'\n')
f.close()
```

```python
f=open('test.txt','r')

while True:
    name=input('이름: ')
    if not name:break
    f.write(name+'\n')
f.close()
```

```python
import os.path

fname=input('파일명: ')
fname=fname+'.txt'
if os.path.exists(fname): # os.path.isfile(fname)
    f=open(fname,'r')
    data=f.read()
    print(data,end='')
    f.close()
else:
    print('{} 파일이 없습니다.'.format(fname))
```

```python
fname=input('파일명: ')
fname=fname+'.txt'
try:
    f=open(fname)
    data=f.read()
    print(data,end='')
    f.close()
except:
    print('{} 파일이 없습니다.'.format(fname))
```
