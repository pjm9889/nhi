###문제3

```python
month = int(input('월 입력: '))
if month < 1 or month > 12 :
    print('존재하지 않는 월!!')
elif 3 <= month <= 5 :
    print(f'{month}월은 봄')
elif 6 <= month <= 8 :
    print(f'{month}월은 여름')
elif 9 <= month <= 11 :
    print(f'{month}월은 가을')
else :
    print(f'{month}월은 겨울')
```

###문제5

```python
import random
com = random.randint(1,30)
print('<< 1~30 숫자 맞히기 게임 >>')
while True :
    player = int(input('숫자 입력(종료 0): '))
    if player == 0 :
        break
    elif player == com :
        print('정답!!')
        break
    elif player > com :
        print('더 작은 숫자 입력! ')
    elif player < com :
        print('더 큰 숫자 입력! ')
```

###문제6

```python
import random
word = ['말했잖아','언젠가', '내버려두지','낭떠러지야','초토화된','불구덩이든','뛰어','낙하', '보여주자','웃자']
input('타자게임 시작 (엔터 입력) ')
w = random.choice(word)
n = 1        # 문제번호
while True :
    print('문제',n,'(종료 0):',w)
    my = input()
    if my == '0' :
        break
    elif my == w :
        print('맞음!!\n')
        w = random.choice(word)
    else :
        print('틀림! 다시!\n')
    n += 1    # 문제번호 증가
```

###문제7

```python
import random
lotto = []

while True:
    num = random.randint(1,45)
    if num not in lotto:
        lotto.append(num)
    if len(lotto)==6:
        break   
print('<< 생성된 로또 번호 >>')
for i in range(6):
    print(f'{lotto[i]} ', end='')
```

###문제8

```python
vote={'대성리':0,'춘천':0,'을왕리':0,'청평':0}
for x in vote:
    print(f'{x}:{vote[x]}표', end=' ')
print('\n')
print('<< MT 장소 투표 >>')
while True:
    area=input('장소: ')
    if not area:break        
    vote[area]=vote[area]+1

for x in vote:
    print(f'{x}:{vote[x]}표', end=' ')

print('\n')
max_key = max(vote, key=vote.get)
print(f'최다득표: {max_key} {vote[max_key]}표')
```

###문제10

```python
def price(menue):
    if menue==1:
       m='아메리카노'
       p=3000
    elif menue==2:
        m='카페라떼'
        p=4000
    elif menue==3:
        m='바닐라라떼'
        p=4500
    print(f'{m}: {p:,}원')

menue=int(input('메뉴선택(1:아메리카노/2:카페라떼/3:바닐라라떼) '))
price(menue)
```
