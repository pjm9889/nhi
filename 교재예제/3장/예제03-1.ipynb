```python
def max1(num1,num2):
    if num1>num2:
        return num1
    else:
      return num2
a=int(input('숫자1: '))
b=int(input('숫자2: '))
print(f'큰수: {max1(a,b)}')
```

```python
def game(num):
    a=num//10
    b=num%10    
    if a==3 or a==6 or a==9:
        return '통과'
    elif b==3 or b==6 or b==9:
        return '통과'
    else:
        return '다음 수'

num=int(input('숫자: '))
print(game(num))
```
