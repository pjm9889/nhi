```python
filename='/content/drive/MyDrive/dust1.xlsx'
dust=pd.read_excel(filename)
dust.head()
```

```python
# 컬럼(변수) 삭제하기
dust=dust.drop(['지역','망','측정소코드'], axis=1)
```

```python
dust.head()
```

```python
# 컬럼생성하기
dust['city']='서울'
```

```python
dust.head()
```

```python
dust.rename(columns={'측정소명':'name',
                     '측정일시':'date',
                     '주소':'addr'}, inplace=True)
```

```python
dust.columns
```

```python
dust.dtypes
```

```python
# 숫자 int형을 문자열형으로 변환
dust['date']=dust['date'].astype(str)
dust['date']=dust['date'].str.slice(0,8)
dust.dtypes
```

```python
dust['date']=pd.to_datetime(dust['date'])
dust.dtypes
```

```python
dust['year']=dust['date'].dt.year
dust['month']=dust['date'].dt.month
dust['day']=dust['date'].dt.day
dust.head()
```

```python
# 문제2
import pandas as pd
filename='/content/drive/MyDrive/data/covid.csv'
df=pd.read_csv(filename,encoding='cp949')
df
#1.항목열을 인덱스열로 지정한 후 ‘스마트폰’의 해당하는 모든 컬럼의 정보를 출력해보자.
df=pd.read_csv(filename,index_col=0,encoding='cp949')
df
#2.사례수가 가장 높은 항목의 모든 정보를 출력해보자.
result=df[df['사례수']==df['사례수'].max()]
result
#3.증가율이 높은 3개의 항목을 출력해보자.
newdata=df.sort_values('증가',ascending=False)
newdata.head(3)
```
