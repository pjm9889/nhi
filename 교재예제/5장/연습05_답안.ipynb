```python
# 문제2
import pandas as pd
filename='/content/drive/MyDrive/data/covid.csv'
df=pd.read_csv(filename,encoding='cp949')
df
#1.항목열을 인덱스열로 지정한 후 ‘스마트폰’의 해당하는 모든 컬럼의 정보를 출력해보자.
df=pd.read_csv(filename,index_col=0,encoding='cp949')
df
#2.사례수가 가장 높은 항목의 모든 정보를 출력해보자.
result=df[df['사례수']==df['사례수'].max()]
result
#3.증가율이 높은 3개의 항목을 출력해보자.
newdata=df.sort_values('증가',ascending=False)
newdata.head(3)
```
